```python
import os

BASE_DIR = "/content/drive/MyDrive/finance_reels"
AUDIO_DIR = f"{BASE_DIR}/audio"
VIDEO_DIR = f"{BASE_DIR}/videos"

os.makedirs(AUDIO_DIR, exist_ok=True)
os.makedirs(VIDEO_DIR, exist_ok=True)

print("Folders ready!")

```

```python
import random

finance_prompts = [
    "Saving money early can make you rich due to compounding.",
    "If you donâ€™t control money, money will control you.",
    "Investing beats saving because inflation eats cash.",
    "Rich people focus on assets, not salary.",
    "Small daily savings create massive wealth over time."
]

def generate_script():
    return random.choice(finance_prompts)

script = generate_script()
script
```

```python
import subprocess
import uuid

def text_to_speech(text):
    audio_file = f"{AUDIO_DIR}/{uuid.uuid4()}.mp3"
    cmd = f'edge-tts --voice "en-IN-NeerjaNeural" --text "{text}" --write-media {audio_file}'
    subprocess.run(cmd, shell=True)
    return audio_file

audio_path = text_to_speech(script)
audio_path
```

```python
import uuid

def create_text_video(text, audio_file):
    output_video = f"{VIDEO_DIR}/{uuid.uuid4()}.mp4"

    ffmpeg_cmd = f"""
    ffmpeg -y \
    -f lavfi -i color=c=black:s=1080x1920:d=6 \
    -i {audio_file} \
    -vf "drawtext=text='{text}':
         fontcolor=white:
         fontsize=64:
         line_spacing=10:
         box=1:
         boxcolor=black@0.6:
         boxborderw=20:
         x=(w-text_w)/2:
         y=(h-text_h)/2,
         zoompan=z='min(zoom+0.0008,1.08)':d=150" \
    -c:v libx264 -c:a aac -shortest {output_video}
    """
    subprocess.run(ffmpeg_cmd, shell=True)
    return output_video

video_path = create_text_video(script, audio_path)
video_path
```

```python
def add_background_music(video, music):
    final_video = video.replace(".mp4", "_final.mp4")
    cmd = f"""
    ffmpeg -y -i {video} -i {music} \
    -filter_complex "amix=inputs=2:weights=2 0.3" \
    -c:v copy {final_video}
    """
    subprocess.run(cmd, shell=True)
    return final_video
```

```python
for i in range(4):
    script = generate_script()
    audio = text_to_speech(script)
    video = create_text_video(script, audio)
    print(f"Generated video {i+1}: {video}")
```
